```
#IMPORTING DIFFERENT MODULES
import numpy as np
import pandas as pd
import pyspark.sql.functions as F
from pyspark.sql import SparkSession
```

```
#CONVERTING THE LOADED DATASET IN PANDAS DATAFRAME FOR NULL HANDLING
#rf-RESULTANT FRAME
rf=df.toPandas()
```

```
#NULL HANDLING
```

```
rf["All_Sky_Surface_PAR_Total"] = pd.to_numeric(rf["All_Sky_Surface_PAR_Total"], downcast="float")
rf['All_Sky_Surface_PAR_Total'] = rf['All_Sky_Surface_PAR_Total'].fillna(rf['All_Sky_Surface_PAR_Total'].mean())
rf['All_Sky_Surface_PAR_Total']
```

```
rf["All_Sky_Surface_UVA_Irradiance"] = pd.to_numeric(rf["All_Sky_Surface_UVA_Irradiance"], downcast="float")
rf['All_Sky_Surface_UVA_Irradiance'] = rf['All_Sky_Surface_UVA_Irradiance'].fillna(rf['All_Sky_Surface_UVA_Irradiance'].mean())
rf['All_Sky_Surface_UVA_Irradiance']
```

```
rf["All_Sky_Surface_UVB_Irradiance"] = pd.to_numeric(rf["All_Sky_Surface_UVB_Irradiance"], downcast="float")
rf['All_Sky_Surface_UVB_Irradiance'] = rf['All_Sky_Surface_UVB_Irradiance'].fillna(rf['All_Sky_Surface_UVB_Irradiance'].mean())
rf['All_Sky_Surface_UVB_Irradiance']
```

```
rf["ALLSKY_SFC_UV_INDEX"] = pd.to_numeric(rf["ALLSKY_SFC_UV_INDEX"], downcast="float")
rf['ALLSKY_SFC_UV_INDEX'] = rf['ALLSKY_SFC_UV_INDEX'].fillna(rf['ALLSKY_SFC_UV_INDEX'].mean())
rf['ALLSKY_SFC_UV_INDEX']
```

```
rf["PRECIPITATION_CORRELATED"] = pd.to_numeric(rf["PRECIPITATION_CORRELATED"], downcast="float")
rf['PRECIPITATION_CORRELATED'] = rf['PRECIPITATION_CORRELATED'].fillna(rf['PRECIPITATION_CORRELATED'].mean())
rf['PRECIPITATION_CORRELATED']
```

```
rf["Earth_Skin_Temperature"] = pd.to_numeric(rf["Earth_Skin_Temperature"], downcast="float")

rf['Earth_Skin_Temperature'] = rf['Earth_Skin_Temperature'].fillna(rf['Earth_Skin_Temperature'].mean())
rf['Earth_Skin_Temperature']
```

```
rf["Surface_Pressure"] = pd.to_numeric(rf["Surface_Pressure"], downcast="float")
rf['Surface_Pressure'] = rf['Surface_Pressure'].fillna(rf['Surface_Pressure'].mean())
rf['Surface_Pressure']
```

```
rf["Relative_Humidity_at_2_Meters"] = pd.to_numeric(rf["Relative_Humidity_at_2_Meters"], downcast="float")
rf['Relative_Humidity_at_2_Meters'] = rf['Relative_Humidity_at_2_Meters'].fillna(rf['Relative_Humidity_at_2_Meters'].mean())
rf['Relative_Humidity_at_2_Meters']
```

```
#DISPLAYING THE MASTER DATASET After NULL HANDLING
display(rf)
```

```
#CONVERTING THE PANDAS DATAFRAME(rf) TO PYSPARK DATAFRAME(newdf)
spark.conf.set("spark.sql.execution.arrow.enabled", "true")
newdf = spark.createDataFrame(rf)
display(newdf)
```

```
#SAVING THE MASTER DATASET AFTER NULL HANDLING IN TABLES FORMAT FOR PART 3 VISUALIZATION
newdf.write.format("delta").mode("overwrite").saveAsTable("default.MASTER_DATASET")
```

```
###ANALYSIS OF 2ND PART###
```

```
#ANALYSING MEAN TEMPERATURE FOR EACH STATE AND ORDERED BY YEAR
temp_avg = newdf.groupby('STATE', 'YEAR').agg(F.avg('Earth_Skin_Temperature').alias('MEAN_TEMPERATURE')).orderBy(F.col('STATE'), F.col('YEAR')).cache()
temp_avg.show()
```

```
#ANALYSING MEAN PRECIPITATION FOR EACH STATE AND ORDERED BY YEAR
pre_avg = newdf.groupby('STATE', 'YEAR').agg(F.avg('PRECIPITATION_CORRELATED').alias('MEAN_PRECIPITATION')).orderBy(F.col('STATE'), F.col('YEAR')).cache()
pre_avg.show()
```

```
#ANALYSING MEAN PRESSURE FOR EACH STATE AND ORDERED BY YEAR
SURFACE_PRESSURE_avg =newdf.groupby('STATE', 'YEAR').agg(F.avg('Surface_Pressure').alias('MEAN_PRESSURE')).orderBy(F.col('STATE'), F.col('YEAR')).cache()
SURFACE_PRESSURE_avg.show()
```

```
#ANALYSING MEAN RELATIVE_HUMIDITY FOR EACH STATE AND ORDERED BY YEAR
Relative_Humidity_avg =newdf.groupby('STATE', 'YEAR').agg(F.avg('Relative_Humidity_at_2_Meters').alias('MEAN_RELATIVE_HUMIDITY_AT_2m')).orderBy(F.col('STATE'), F.col('YEAR'))
Relative_Humidity_avg.show()
```

```
###SAVING INTO TABLES FOR ANALYSIS OF PART 2###
Relative_Humidity_avg.write.format("delta").mode("overwrite").saveAsTable("default.RELATIVE_HUMIDITY")
SURFACE_PRESSURE_avg.write.format("delta").mode("overwrite").saveAsTable("default.SURFACE_PRESSURE")
pre_avg.write.format("delta").mode("overwrite").saveAsTable("default.PRECIPITATION")
temp_avg .write.format("delta").mode("overwrite").saveAsTable("default.TEMPERATURE")
```
